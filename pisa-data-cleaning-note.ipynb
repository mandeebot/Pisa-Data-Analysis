DATA CLEANING

```python
#feature encoding
# using points system to encode the scale used for the categorical features 
scale_1={'Strongly agree':4,'Agree': 3,'Disagree': 2, 'Strongly disagree':1}
scale_2={'Always or almost always': 4, 'Often':3, 'Sometimes':2, 'Never or rarely':1}
scale_3={'Not at all like me':5,'Not much like me':4,'Somewhat like me':3,'Mostly like me':2, 'Very much like me' :1}


# Function that will assign point values to each feature category
def scale_point(col_names,f_name,scale):
    for col in col_names:
        data.loc[:,col]=data.loc[:,col].map(scale)
    data[f_name]=data.loc[:,col_names].sum(axis=1, skipna=False)/len(col_names)
```

```python
#encoding catgerorical feature
# Motivation
scale_point(motivation1,'motivation',scale_1)

# Interest
scale_point(interest1, 'interest', scale_1)

# Behaivior
scale_point(behaivior1, 'behaivior', scale_2)

#Problem Solving Capability
scale_point(prob_solv_skills1, 'prob_solv_skills', scale_3)
```
